```python
#extracts gene of interest
df_fpkms = pd.read_csv( "../../qbb2020/data/fpkms.csv", index_col="t_name" )
goi = "FBtr0331261"
df_sxl = df_fpkms.loc[goi,:]
x = df_sxl[1:]
male=x[:8]
female=x[8:]
x
```

```python
fig, ax = plt.subplots()
index=["10","11","12","13","14A","14B","14C","14D"]
ax.plot(index , male,label="male" )
ax.plot(index,female,label="female")
ax.legend()
plt.show()
```
