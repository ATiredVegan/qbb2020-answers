# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Processing Data

```python
def process_bed_freq(filename,bed_columns):
    df_k4 = pd.read_table(filename, names=bed_columns )
    df_k4=df_k4.sort_values(by="chr")
    #extracts only chromosome data
    chr_k4 = df_k4.loc[:,"chr"]
    # count each chromosome, sorting by existing index
    chr_k4 = chr_k4.value_counts().sort_index()
    return chr_k4
def process_bed_width(filename,bed_columns):
    df_k4 = pd.read_table(filename, names=bed_columns )
    df_k4.loc[:,"width"] = df_k4.loc[:,"end"] - df_k4.loc[:,"start"]
    chr_k4_width = df_k4.loc[:,"width"]
    # count each chromosome, sorting by existing index
    
    chr_k4_width = chr_k4_width.value_counts().sort_index()

    return chr_k4_width
```

```python
bed_columns = ["chr", "start", "end"]
chr_k4=process_bed_freq("../../qbb2020/data/K4me3.bed",bed_columns)
chr_k9=process_bed_freq("../../qbb2020/data/K9me3.bed",bed_columns)
chr_k27=process_bed_freq("../../qbb2020/data/K27me3.bed",bed_columns)
chr_k4_width=process_bed_width("../../qbb2020/data/K4me3.bed",bed_columns)
chr_k9_width=process_bed_width("../../qbb2020/data/K9me3.bed",bed_columns)
chr_k27_width=process_bed_width("../../qbb2020/data/K27me3.bed",bed_columns)
```

# Plot Data

```python
#plot 
fig, ax = plt.subplots(figsize=(10,8))
ax.bar( chr_k4.index, chr_k4 )
ax.set_title("Chrosome Freequency of K3 methylation in Fruit Flies")
ax.set_xlabel("Chromosome")
ax.set_ylabel("Frequency")
plt.show()
```

# Question 3

```python
# plots data together 
fig, ax = plt.subplots(ncols=3, figsize=(30,8), sharey=True)
ax[0].bar( chr_k4.index, chr_k4 )
ax[1].bar(chr_k9.index, chr_k9)
ax[2].bar(chr_k27.index, chr_k27)
ax[0].set_ylabel("Frequency")
ax[0].set_xlabel("K4 methylation")
ax[1].set_xlabel("K9 methylation")
ax[2].set_xlabel("K27 methylation")
fig.suptitle("Chromsome Methylation Frequency in Fruit Flies")
plt.show()
```

```python
# plots data together 
fig, ax = plt.subplots(ncols=3, figsize=(30,8), sharey=False)
ax[0].hist(chr_k4_width)
ax[1].hist(chr_k9_width)
ax[2].hist(chr_k27_width)
ax[0].set_ylabel("Frequency of Width")
ax[0].set_xlabel("K4 methylation width")
ax[1].set_xlabel("K9 methylation width")
ax[2].set_xlabel("K27 methylation width")
fig.suptitle("Chromsome Methylation Width in Fruit Flies")
plt.show()
```
