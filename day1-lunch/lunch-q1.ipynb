```python
#imports 
import matplotlib.pyplot as plt
```

```python
# initialize lists
x=[1,2,3,4,5]
y = [1, 4, 9, 16, 25]
x2 = [2, 4, 6]
y2 = [8, 64, 216]
```

```python
#plot 

fig, ax = plt.subplots(nrows=2,sharex=True, sharey=False)
ax[0].plot( x, y, '.')
ax[1].plot( x2, y2)
ax[0].get_shared_x_axes().join(ax[0],ax[1])
ax[1].set_xlabel("x")
ax[0].set_title("Squares and Cubes")
ax[0].set_ylabel("x^2")
ax[1].set_ylabel("x^3")
plt.show()
```
