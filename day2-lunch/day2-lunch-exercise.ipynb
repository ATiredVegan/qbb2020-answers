# Part 1: 
Import and create a new file with only the FlyBase ID and UniProt

```python
#opens the file to read genes from and a new file to write only FLyBase and UniProt
fly_genes=open('fly.txt','r')
uniprot=open('cleangenes.txt','w+')
#writes headers
uniprot.write('Uniprot \t FlyBase')
#iterates through every gene
for line in fly_genes:
    line=line.split()
    #ensures that the line is not a header, and the Swiss Prot contains the DROME identifier
    if len(line)>1 and 'DROME' in line[1] and 'FB' in line[-1]:
        uniprot.write('\n'+line[2]+'\t'+line[3])
fly_genes.close()
uniprot.close()
        
```

# Part 2 
Function that replaces gene id. We also have two optional variables for users that say if they want to include entries with no uniprotID. 

```python
def identifier_mapper(mapping_filename,c_tab,deleteNoID=True,missingname="NoUniProtID"): 
    uniprot_mapping=open(mapping_filename,'r')
    ctab=open(c_tab,'r')
    uniprot_dict={}
    for line in uniprot_mapping:
        line=line.split('\t')
        #-1 removes newline character
        uniprot_dict[line[1][:-1]]=line[0] 
    counter=0
    for line in ctab:
        if 't_id' in line:
            print(line)
        else:
            line=line.split('\t')
            if line[8] in uniprot_dict:
                line[8]=uniprot_dict[line[8]]
                counter+=1
                line[0]=str(counter)
            elif not deleteNoID:
                line[8]=missingname
                counter+=1
                line[0]=str(counter)
            else:
                continue
            a="".join(elem+"\t" for elem in line)
            print(a)
        if counter>=100:
            break
    
```

```python
identifier_mapper('cleangenes.txt','t_data.ctab')
```
